## Question 1: 

Select a stock and create a suitable plot for it. Make sure the plot is readable with relevant information, such as date, values.  

```python
# YOUR CODE HERE
#plt.plot(stocks['date'],stocks['GOOG'])
#plt.xlabel("date")
#plt.ylabel("stock value")
#plt.xticks(stocks['date'][::25])
#plt.title("Google stock")
#plt.show()

x = stocks['date']
y = stocks['GOOG']
fig,ax = plt.subplots(dpi = 150)
ax.plot(x,y,marker = '.')
ax.set_xlabel('Date')
ax.set_ylabel('Stock Value')
ax.set_title('Google Stock')
ax.set_xticks(x[::20])
for i in range(0,len(x),2):
    ax.text(x[i],y[i],'%.2f' %y[i],fontsize = 7)
plt.show()
```

## Question 2: 

You've already plot data from one stock. It is possible to plot multiples of them to support comparison.  
To highlight different lines, customise line styles, markers, colors and include a legend to the plot.

```python
# YOUR CODE HERE
x = stocks['date']
y = stocks.columns[1:]
#print(y)
markers = ['.',',','o','v','^','<']
colors = ['blue','orange','green','red','purple','brown']
linestyles = ['-','--',':','-.',':','-']
fig,ax = plt.subplots(dpi=150)
j=0
for i in y:
    ax.plot(x,stocks[i],marker = markers[j],markersize = 3,color = colors[j],linestyle = linestyles[j])
    j+=1
ax.set_xlabel('Date')
ax.set_xticks(ax.get_xticks()[::20])
ax.set_ylabel('Stock Value')
ax.set_title('Stocks')
ax.legend(y)

plt.show()
```

## Question 5:

Recreate the barplot below that shows the population of different continents for the year 2007. 

*Hints:*

```python
#load data
df = px.data.gapminder()
df.head()
```

```python
# YOUR CODE HERE
df_sum = df.groupby('continent').sum()
fig = px.bar(df_sum,x='pop',y=df_sum.index,orientation='h',color=df_sum.index,text='pop')
fig.update_yaxes(categoryorder='total ascending')
fig.show()
```
